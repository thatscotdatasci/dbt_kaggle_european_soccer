```python
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sns; sns.set()
```

```python
sql_team_data = 'SELECT * FROM public_dbt_teams.team_data'
sql_player_data = 'SELECT * FROM public_dbt_players.player_data'
```

```python
engine = create_engine('postgresql+psycopg2://username:password@localhost:5432/database', echo=False)
connection = engine.connect()
```

```python
team_data = pd.read_sql(sql_team_data, connection)
```

```python
print(team_data.head())
```

```python
team_scatter = sns.scatterplot(x='avg_age', y='win', hue='bmi_cat', size='avg_bmi', data=team_data)
```

```python
player_data = pd.read_sql(sql_player_data, connection)
```

```python
print(player_data.head())
```

```python
player_scatter_bmi = sns.scatterplot(x='age', y='win', hue='bmi_cat', size='bmi', data=player_data)
```

```python
player_scatter_career_length = sns.scatterplot(x='career_length', y='win', data=player_data)
```
